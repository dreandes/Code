```python
data = sm.datasets.get_rdataset("Diamonds", package = "Stat2Data")
df = data.data
df.tail()
```

```python
model = sm.OLS.from_formula("TotalPrice ~ Carat", data=df)
result = model.fit()
print(result.summary())
```

```python
fig = sm.graphics.plot_regress_exog(result, "Carat")
plt.tight_layout(pad=4, h_pad=0.5, w_pad=0.5)
plt.show()
```

```python
model = sm.OLS.from_formula("TotalPrice ~ Carat + I(Carat**2)", data=df)
result = model.fit()
print(result.summary())
```

```python
fig = sm.graphics.plot_regress_exog(result, "I(Carat ** 2)")
plt.tight_layout(pad=4, h_pad=0.5, w_pad=0.5)
plt.show()
```

```python
model = sm.OLS.from_formula("TotalPrice ~ Carat + I(Carat**2) + Depth", data=df)
result = model.fit()
print(result.summary())
```

```python
fig = sm.graphics.plot_regress_exog(result, "Depth")
plt.tight_layout(pad=4, h_pad=0.5, w_pad=0.5)
plt.show()
```

```python
model = sm.OLS.from_formula("TotalPrice ~ Carat + I(Carat**2) + Depth + C(Color) + 0", data=df)
result = model.fit()
print(result.summary())
```

```python
model = sm.OLS.from_formula("TotalPrice ~ Carat + I(Carat**2) + Depth + C(Color) + C(Clarity)", data=df)
result = model.fit()
print(result.summary())
```

```python
model = sm.OLS.from_formula("TotalPrice ~ scale(Carat) + scale(I(Carat**2)) + scale(Depth) + C(Color) + 0", data=df)
result = model.fit()
print(result.summary())
```
