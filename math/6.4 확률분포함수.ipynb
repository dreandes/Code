```python
t = np.linspace(-100, 500, 100)
F = t / 360
F[t < 0] = 0
F[t > 360] = 1
plt.plot(t, F)
plt.ylim(-0.1, 1.1)
plt.xticks([-100, 0, 90, 180, 270, 360, 500])
plt.title("누적분포함수")
plt.xlabel("$x$ (도)")
plt.ylabel("$F(x)$")
plt.show()
```

```python
def F(a):
    if a < 0:
        return 0
    if a > 360:
        return 1
    elif a < 180:
        return 2/3 * a/180
    elif a >= 180:
        return 2/3 + 1/3 * (a-180)/180

t = np.linspace(-100, 500, 100)
Fs = np.array(list(map(F, t)))
plt.plot(t, Fs)
plt.ylim(-0.1, 1.1)
plt.xticks([-100, 0, 90, 180, 270, 360, 500])
plt.title("누적분포함수")
plt.xlabel("$x$ (도)")
plt.ylabel("$F(x)$")
plt.show()
```
